```python
# Calculate rolling volatility (3-year window) for each occupation
volatility_analysis = {}
recovery_analysis = {}

for occ_code in percent_of_market.columns:
    if occ_code == 'Total_OCC_Sum':
        continue
    
    # Calculate YoY % change
    yoy_change = percent_of_market[occ_code].pct_change() * 100
    
    # Calculate rolling volatility (3-year std dev)
    rolling_volatility = yoy_change.rolling(window=3).std()
    
    volatility_analysis[occ_code] = {
        'Overall_Volatility': yoy_change.std(),
        'Rolling_Volatility_Mean': rolling_volatility.mean(),
        'Max_Rolling_Volatility': rolling_volatility.max(),
        'Resilience_Score': 100 / (yoy_change.std() + 0.1)  # Higher score = more resilient
    }
    
    # Calculate recovery time for 2008 crisis specifically
    pre_crisis_level = percent_of_market[occ_code].loc['2007'].mean()
    min_level_during_crisis = percent_of_market[occ_code].loc['2008':'2009'].min()
    
    # Find first year after crisis when employment returns to pre-crisis level
    recovered = False
    recovery_years = 0
    for year in range(2010, 2025):
        year_data = percent_of_market[occ_code].loc[str(year)].mean()
        if year_data >= pre_crisis_level * 0.95:  # 95% threshold for recovery
            recovery_years = year - 2008
            recovered = True
            break
    
    recovery_analysis[occ_code] = {
        'Pre_Crisis_Level': pre_crisis_level,
        'Min_Crisis_Level': min_level_during_crisis,
        'Shock_Magnitude_%': ((min_level_during_crisis - pre_crisis_level) / pre_crisis_level * 100),
        'Recovery_Years': recovery_years if recovered else 15,  # Cap at 15 if not recovered
        'Status': 'Recovered' if recovered else 'Still Below Pre-Crisis'
    }

# Create DataFrames
volatility_df = pd.DataFrame(volatility_analysis).T
volatility_df = volatility_df.sort_values('Resilience_Score', ascending=False)

recovery_df = pd.DataFrame(recovery_analysis).T
recovery_df = recovery_df.sort_values('Recovery_Years')

# Map OCC codes to readable titles
volatility_df.index = volatility_df.index.map(lambda x: occ_code_to_title.get(x, x))
recovery_df.index = recovery_df.index.map(lambda x: occ_code_to_title.get(x, x))

print("=" * 100)
print("EMPLOYMENT RESILIENCE RANKINGS (Based on Volatility)")
print("=" * 100)
print("\nHigh Resilience (Low Volatility) = Stable Sector:")
print(volatility_df[['Overall_Volatility', 'Resilience_Score']].head(10))

print("\n\nLow Resilience (High Volatility) = Unstable Sector:")
print(volatility_df[['Overall_Volatility', 'Resilience_Score']].tail(10))

print("\n" + "=" * 100)
print("2008 FINANCIAL CRISIS: RECOVERY ANALYSIS")
print("=" * 100)
print("\nSectors ranked by recovery speed (fastest recovery first):")
print(recovery_df[['Shock_Magnitude_%', 'Recovery_Years', 'Status']].head(15))
```

```python
# Calculate year-over-year employment change for all major occupations
yoy_employment_change = percent_of_market.pct_change() * 100

# Create event period classifications
def classify_event_period(year):
    if 1999 <= year < 2000:
        return 'Pre-Dot-Com'
    elif 2000 <= year <= 2002:
        return 'Dot-Com Crisis'
    elif 2003 <= year < 2007:
        return 'Recovery (Post-Dot-Com)'
    elif 2007 <= year <= 2009:
        return '2008 Financial Crisis'
    elif 2010 <= year < 2020:
        return 'Recovery (Post-2008)'
    elif 2020 <= year <= 2021:
        return 'COVID-19 Pandemic'
    else:
        return 'Post-COVID Recovery'

df['Event_Period'] = df['Year_int'].apply(classify_event_period)

# Calculate employment statistics by event period and occupation
event_analysis = df[df['O_GROUP'] == 'major'].groupby(['Event_Period', 'OCC_CODE_CLEANED']).agg({
    'TOT_EMP': ['mean', 'min', 'max', 'std'],
    'REAL_A_MEDIAN': ['mean', 'min', 'max']
}).round(0)

print("=" * 80)
print("EMPLOYMENT STATISTICS BY EVENT PERIOD")
print("=" * 80)
print("\nNote: These show average employment (in absolute numbers) and wage statistics by period")
print("\nKey findings by event period:")

# Calculate shock magnitude for each event
events = {
    'Dot-Com (2000-2002)': (2000, 2002),
    '2008 Crisis (2007-2009)': (2007, 2009),
    'COVID-19 (2020-2021)': (2020, 2021)
}

shock_results = []
for event_name, (start_yr, end_yr) in events.items():
    pre_year = start_yr - 1
    
    pre_data = percent_of_market.loc[str(pre_year)].squeeze()
    during_data = percent_of_market.loc[str(start_yr):str(end_yr)].min().squeeze()

    shock_magnitude = ((during_data - pre_data) / pre_data * 100).sort_values()
    
    shock_results.append({
        'Event': event_name,
        'Pre-Crisis Year': pre_year,
        'Crisis Years': f"{start_yr}-{end_yr}",
        'Most_Impacted_Sector': shock_magnitude.index[0],
        'Max_Shock_Magnitude_%': shock_magnitude.iloc[0]
    })

shock_df = pd.DataFrame(shock_results)
display(shock_df)

# Detailed look at most impacted sectors during each crisis
print("\n" + "=" * 80)
print("SECTOR-SPECIFIC SHOCK IMPACT ANALYSIS")
print("=" * 80)

for event_name, (start_yr, end_yr) in events.items():
    pre_year = start_yr - 1
    pre_data = percent_of_market.loc[str(pre_year)].squeeze()
    during_data = percent_of_market.loc[str(start_yr):str(end_yr)].min().squeeze()
    
    shock_pct_change = ((during_data - pre_data) / pre_data * 100).sort_values()
    
    print(f"\n{event_name}:")
    print(f"  Most Resilient (smallest decline): {shock_pct_change.index[-1]} ({shock_pct_change.iloc[-1]:.2f}%)")
    print(f"  Most Vulnerable (largest decline): {shock_pct_change.index[0]} ({shock_pct_change.iloc[0]:.2f}%)")
    print(f"  Average sector decline: {shock_pct_change.mean():.2f}%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Configure plotting
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Load data
df = pd.read_excel('national_bls_stats_cleaned.xlsx')
cpi_df = pd.read_excel('CPI-U_index_monthly.xlsx')

# Data cleaning (consistent with original notebook)
df = df.drop(columns=['JOBS_1000','OCC_CODE', 'LOC_QUOTIENT', 'PCT_TOTAL', 'PCT_RPT'])
single_value_cols = ['AREA', 'AREA_TITLE', 'AREA_TYPE', 'PRIM_STATE', 'NAICS', 'NAICS_TITLE', 'I_GROUP', 'OWN_CODE']
df = df.drop(columns=single_value_cols)

# Data type conversion
df['ANNUAL'] = df['ANNUAL'].astype(str)
df['HOURLY'] = df['HOURLY'].astype(str)
df['OCC_CODE_CLEANED'] = df['OCC_CODE_CLEANED'].astype(str)
df['OCC_TITLE'] = df['OCC_TITLE'].astype(str)
df['O_GROUP'] = df['O_GROUP'].astype(str)
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

numeric_cols = [col for col in df.columns if col not in ['ANNUAL', 'HOURLY', 'OCC_CODE_CLEANED', 'OCC_TITLE', 'O_GROUP', 'Year']]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# CPI merge for inflation adjustment
cpi_df['Year'] = pd.to_datetime(cpi_df['Year'], format='%Y')
cpi_join = cpi_df[['Year', 'Annual']].rename(columns={'Annual': 'CPI'})
df = df.merge(cpi_join, on='Year', how='left')

current_CPI = df[df['Year']=='2024-01-01']['CPI'].values[0]
df['REAL_A_MEAN'] = df['A_MEAN'] * (current_CPI / df['CPI'])
df['REAL_A_MEDIAN'] = df['A_MEDIAN'] * (current_CPI / df['CPI'])

# Feature engineering
df['OCC_2DIGIT'] = df['OCC_CODE_CLEANED'].str[:2]
df['Year_int'] = df['Year'].dt.year

# Get occupation mappings
occ_code_to_title = df[['OCC_CODE_CLEANED', 'OCC_TITLE']].drop_duplicates().set_index('OCC_CODE_CLEANED')['OCC_TITLE'].to_dict()

# Create market share data
broad_df = df[(df['O_GROUP'] == 'major')][['Year', 'OCC_CODE_CLEANED','TOT_EMP']].copy()
broad_df = broad_df[broad_df['OCC_CODE_CLEANED'] != '00-0000'].copy()

contingency_table = pd.crosstab(index=broad_df['Year'], columns=broad_df['OCC_CODE_CLEANED'],values=broad_df['TOT_EMP'],aggfunc='sum')
contingency_table_scaled = contingency_table / 1000000
percent_of_market = contingency_table_scaled.div(contingency_table_scaled.sum(axis=1), axis=0) * 100

print("Data loaded and processed successfully")
print(f"Date range: {df['Year'].min()} to {df['Year'].max()}")
print(f"Total occupation codes: {df['OCC_CODE_CLEANED'].nunique()}")
print(f"Major occupation groups: {len([x for x in df[df['O_GROUP']=='major']['OCC_CODE_CLEANED'].unique() if x != '00-0000'])}")
```
